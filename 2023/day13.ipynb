```python
import numpy as np
def day_13_part_01(arr):
    result = 0
    for i, pattern in enumerate(arr):
        indexOne = [[], []]
        indexTwo = [[], []]
        for j in range(1, len(pattern)):
            if pattern[j] == pattern[j - 1]:
                indexOne[0].append(j - 1)
                indexOne[1].append(j - 1)
                indexTwo[0].append(j)
                indexTwo[1].append(j)
        
        partialResult = [False]
        if len(indexOne[0]) != 0:
            for k, _ in enumerate(indexOne[0]):
                onGoing = True
                while True:
                    if pattern[indexOne[1][k]] != pattern[indexTwo[1][k]]:
                        break
                    indexOne[1][k] -= 1
                    indexTwo[1][k] += 1
                    if indexOne[1][k] == -1 or indexTwo[1][k] == len(pattern):
                        onGoing = False
                        partialResult = [True, indexTwo[0][k]]
                        break
        
        if partialResult[0]:
            result += (partialResult[1] * 100)
        else:
            newPattern = []
            for i, _ in enumerate(pattern[0]):
                newRow = ''
                for _, row in enumerate(pattern):
                    newRow += row[i]
                newPattern.append(newRow)
            
            indexThree = [[], []]
            indexFour = [[], []]
            for j in range(1, len(newPattern)):
                if newPattern[j] == newPattern[j - 1]:
                    indexThree[0].append(j - 1)
                    indexThree[1].append(j - 1)
                    indexFour[0].append(j)
                    indexFour[1].append(j)
            
            partialResult = [False]
            if len(indexThree[0]) != 0:
                for k, _ in enumerate(indexThree[0]):
                    onGoing = True
                    while onGoing:
                        if newPattern[indexThree[1][k]] != newPattern[indexFour[1][k]]:
                            break
                        indexThree[1][k] -= 1
                        indexFour[1][k] += 1
                        if indexThree[1][k] == -1 or indexFour[1][k] == len(newPattern):
                            onGoing = False
                            partialResult = [True, indexFour[0][k]]
                            break

            if partialResult[0]:
                result += partialResult[1]

    print(result)
```

```python
def day_13_part_02(arr):
    def count_differences(arr):
        diff = 0
        for i, char in enumerate(arr[0]):
            if not char == arr[1][i]:
                diff += 1
        return diff

    result = 0
    for i, pattern in enumerate(arr):
        indexOne = [[], []]
        indexTwo = [[], []]
        for j in range(1, len(pattern)):
            if count_differences([pattern[j], pattern[j - 1]]) in [0, 1]:
                indexOne[0].append(j - 1)
                indexOne[1].append(j - 1)
                indexTwo[0].append(j)
                indexTwo[1].append(j)
        
        partialResult = [False]
        if len(indexOne[0]) != 0:
            for k, _ in enumerate(indexOne[0]):
                diff = 0
                while True:
                    diff += count_differences([pattern[indexOne[1][k]], pattern[indexTwo[1][k]]])

                    indexOne[1][k] -= 1
                    indexTwo[1][k] += 1
                    if indexOne[1][k] == -1 or indexTwo[1][k] == len(pattern):
                        if diff == 1:
                            partialResult = [True, indexTwo[0][k]]
                        break
        
        if partialResult[0]:
            result += (partialResult[1] * 100)
        else:
            newPattern = []
            for i, _ in enumerate(pattern[0]):
                newRow = ''
                for _, row in enumerate(pattern):
                    newRow += row[i]
                newPattern.append(newRow)
            
            indexThree = [[], []]
            indexFour = [[], []]
            for j in range(1, len(newPattern)):
                if count_differences([newPattern[j], newPattern[j - 1]]) in [0, 1]:
                    indexThree[0].append(j - 1)
                    indexThree[1].append(j - 1)
                    indexFour[0].append(j)
                    indexFour[1].append(j)
            
            partialResult = [False]
            if len(indexThree[0]) != 0:
                for k, _ in enumerate(indexThree[0]):
                    diff = 0
                    while True:
                        diff += count_differences([newPattern[indexThree[1][k]], newPattern[indexFour[1][k]]])

                        indexThree[1][k] -= 1
                        indexFour[1][k] += 1
                        if indexThree[1][k] == -1 or indexFour[1][k] == len(newPattern):
                            if diff == 1:
                                partialResult = [True, indexFour[0][k]]
                            break

            if partialResult[0]:
                result += partialResult[1]

    print(result)
```
